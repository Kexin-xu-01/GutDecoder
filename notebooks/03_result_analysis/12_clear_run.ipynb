clear incomplete runs (those without dataset_results.csv)

```python
import os
import shutil
import time


# Root directory containing the run folders
ROOT_DIR = "/project/simmons_hts/kxu/hest/eval/ST_pred_results"
EXPECTED_FILE = "dataset_results.csv"
TIME_THRESHOLD_HOURS = 24


# Current time
now = time.time()


# Loop through all run directories
for folder in os.listdir(ROOT_DIR):
    run_path = os.path.join(ROOT_DIR, folder)


    if os.path.isdir(run_path) and folder.startswith("run_"):
        result_file = os.path.join(run_path, EXPECTED_FILE)
        last_modified = os.path.getmtime(run_path)
        hours_since_mod = (now - last_modified) / 3600


    # If expected file is missing and folder not modified in the past 24 hours, delete the folder
    if not os.path.isfile(result_file) and hours_since_mod > TIME_THRESHOLD_HOURS:
        print(f"[REMOVING] {run_path} (missing {EXPECTED_FILE} and not modified in last 24h)")
        shutil.rmtree(run_path)
    else:
        print(f"[OK] {run_path}")
```
