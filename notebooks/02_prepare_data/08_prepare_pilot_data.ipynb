```python
from hest import iter_hest
from hest.HESTData import create_splits
import pandas as pd
from hest.utils import get_k_genes
from typing import List
import os
import shutil
```

```python
meta = pd.read_csv("/project/simmons_hts/kxu/hest/hest_directory.csv")
meta
```

```python
meta_df = pd.DataFrame()
meta_df['id'] = meta['SampleID']
meta_df['patient'] = meta['Patient ID']
meta_df['dataset_title'] = meta_df['id'].str.split("ROI").str[0]
meta_df
```

```python
def get_path_from_meta_row(row, hest_data_dir="/project/simmons_hts/kxu/hest/data"):
    """
    Get base path for sample files in local HEST data.
    Assumes structure:
        st/{id}.h5ad
        patches/{id}.h5
        patches_vis/{id}.png
    """
    sample_id = row["id"]

    # Return base folder (no "processed" subdir)
    return os.path.join(hest_data_dir, sample_id)

def get_k_genes_from_df(meta_df: pd.DataFrame,  k: int, criteria: str, save_dir: str=None,hest_data_dir="/project/simmons_hts/kxu/hest/data") -> List[str]:
    """Get the k genes according to some criteria across common genes in all the samples in the HEST meta dataframe

    Args:
        meta_df (pd.DataFrame): HEST meta dataframe
        k (int): number of genes to return
        criteria (str): criteria for the k genes to return
            - 'mean': return the k genes with the largest mean expression across samples
            - 'var': return the k genes with the largest expression variance across samples
        save_dir (str, optional): genes are saved as json array to this path if not None. Defaults to None.

    Returns:
        List[str]: k genes according to the criteria
    """
    import scanpy as sc
    
    adata_list = []
    for _, row in meta_df.iterrows():
        sample_id = row["id"]
        path = os.path.join(hest_data_dir, "st", f"{sample_id}.h5ad")
        if not os.path.exists(path):
            raise FileNotFoundError(f"Missing file: {path}")
        adata = sc.read_h5ad(path)
        adata_list.append(adata)
    return get_k_genes(adata_list, k, criteria, save_dir=save_dir)
```

```python
import os
import shutil
import pandas as pd

def create_benchmark_data(meta_df, save_dir: str, K: int, hest_data_dir="/project/simmons_hts/kxu/hest/data"):
    """
    Create HEST benchmark data from local data folder.

    Args:
        meta_df (pd.DataFrame): must contain ['id', 'patient', 'dataset_title']
        save_dir (str): output benchmark directory
        K (int): number of splits
        hest_data_dir (str): base folder containing st/, patches/, patches_vis/
    """
    os.makedirs(save_dir, exist_ok=True)

    # fill missing patients if any
    meta_df['patient'] = meta_df['patient'].fillna('Patient 1')

    # generate var_50genes.json
    get_k_genes_from_df(meta_df, 50, 'var', os.path.join(save_dir, 'var_50genes.json'),hest_data_dir)

    # group into splits
    splits = meta_df.groupby(['dataset_title', 'patient'])['id'].agg(list).to_dict()
    create_splits(os.path.join(save_dir, 'splits'), splits, K=K)

    # prepare subfolders
    os.makedirs(os.path.join(save_dir, 'patches'), exist_ok=True)
    os.makedirs(os.path.join(save_dir, 'patches/vis'), exist_ok=True)
    os.makedirs(os.path.join(save_dir, 'adata'), exist_ok=True)

    # copy files for each sample
    for _, row in meta_df.iterrows():
        sample_id = row['id']

        # source files from hest_data_dir
        src_patch = os.path.join(hest_data_dir, "patches", f"{sample_id}.h5")
        dst_patch = os.path.join(save_dir, "patches", f"{sample_id}.h5")
        shutil.copy(src_patch, dst_patch)

        src_vis = os.path.join(hest_data_dir, "patches_vis", f"{sample_id}_patch_vis.png")
        dst_vis = os.path.join(save_dir, "patches/vis", f"{sample_id}.png")
        shutil.copy(src_vis, dst_vis)

        src_adata = os.path.join(hest_data_dir, "st", f"{sample_id}.h5ad")
        dst_adata = os.path.join(save_dir, "adata", f"{sample_id}.h5ad")
        shutil.copy(src_adata, dst_adata)

    print(f"✅ Benchmark dataset created at {save_dir}")
```

```python
create_benchmark_data(
    meta_df,
    save_dir="/project/simmons_hts/kxu/hest/eval/data/XeniumPR1",
    K=15,
    hest_data_dir="/project/simmons_hts/kxu/hest/data"
)
```
